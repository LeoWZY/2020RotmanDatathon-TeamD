```python
import pandas as pd

df  = pd.read_excel("crypto_price_raw.xlsx")
```

```python
import numpy as np
df['price_change_abs'] = df['close'] - df['open']
df['price_change_pcg'] = (df['close'] - df['open']) / df['open'] * 100
df['candle_body_size'] = abs(df['close'] - df['open'])
df['upper_wick_size'] = df['high'] - df[['open', 'close']].max(axis=1)
df['lower_wick_size'] = df[['open','close']].min(axis=1) - df['low']
df['candle_direction'] = np.where(df['close'] > df['open'], 1, 0)
```

```python
# Convert to datetime if not already
df['create_date'] = pd.to_datetime(df['create_date'])

# Extract just the date part
df['date'] = df['create_date'].dt.date
```

```python
df = df.sort_values(['name', 'create_date'])
```

```python
def calculate_metrics(group):

  group['previous_close'] = group['close'].shift(1)

# Calculate 'true_range'
  group['true_range'] = group.apply(
    lambda row: max(
        row['high'] - row['low'],
        abs(row['high'] - row['previous_close']),
        abs(row['low'] - row['previous_close'])
    ),
    axis=1
  )

  atr_period = 14
  group['ATR_14'] = group['true_range'].rolling(window=atr_period).mean()
  group['log_returns'] = np.log(group['close'] / group['close'].shift(1))

  volatility_window = 14
  group['historical_volatility'] = group['log_returns'].rolling(window=volatility_window).std() * np.sqrt(volatility_window)  # Annualize if desired

  #SMA
  group['10_period_SMA'] = group['close'].rolling(window=10).mean()
  group['30_period_SMA'] = group['close'].rolling(window=30).mean()
  group['50_period_SMA'] = group['close'].rolling(window=50).mean()
  group['100_period_SMA'] = group['close'].rolling(window=100).mean()

  #EMA
  group['10_period_EMA'] = group['close'].rolling(window=10).mean()
  group['30_period_EMA'] = group['close'].rolling(window=30).mean()
  group['50_period_EMA'] = group['close'].rolling(window=50).mean()
  group['100_period_EMA'] = group['close'].rolling(window=100).mean()

  ##RSI
  delta=group['close'].diff()
  gain = (delta.where(delta>0,0)).rolling(window=14).mean()
  loss=(-delta.where(delta <0,0)).rolling(window=14).mean()
  rs = gain/loss
  group['RSI_14'] = 100 - (100 /(1+rs))


  ##bollinger bands
  sma_20 = group['close'].rolling(window=20).mean()
  rolling_std = group['close'].rolling(window=20).std()
  group['upper_bb_20'] = sma_20 + (rolling_std *2)
  group['lower_bb_20'] = sma_20 - (rolling_std *2)
  group['bb_middle'] = sma_20

  # ===== VOLUME-BASED METRICS =====
  # Absolute volume changes
  group['token_volume_abs'] = group['token_volume'].diff()
  group['usdt_volume_abs'] = group['usdt_volume'].diff()

  # Percentage volume changes
  group['token_volume_pct'] = (group['token_volume_abs'] / group['token_volume'].shift(1)) * 100
  group['usdt_volume_pct'] = (group['usdt_volume_abs'] / group['usdt_volume'].shift(1)) * 100

  # Volume moving averages (fixed window to 14 per your variable name)
  group['token_volume_moving_avg_14'] = group['token_volume'].rolling(window=14).mean()
  group['usdt_volume_moving_avg_14'] = group['usdt_volume'].rolling(window=14).mean()

  return group

df = df.groupby('name', group_keys=False).apply(calculate_metrics)
```

```python
df.to_csv("crypto_data_hourly_v3.csv")
```
